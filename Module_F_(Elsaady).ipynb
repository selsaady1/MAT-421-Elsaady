Finite Difference Approximation: UtilizesÂ values from the neighborhood of the function to estimate the objective.

Forward Difference - Slope estimation of x_j using a line that connects (x_j, f(x_j)) &(x_jplus1, f(x_jplus1)) f'(x_j) = (f(x_jplus1) - f(x_j)) / (x_jplus1 - x_j)

Backward Difference - Slope of x_j is estimated using a line that connects (x_j, f(x_j)) & (x_jminus1, f(x_minus1)) f'(x_j) = (f(x_j) - f(x_jminus1)) / (x_j - x_jminus1)

Central Difference - The slope of x_j is estimated using line that connects (x_jminus1, f(x_minus1)) & (x_jplus1, f(x_jplus1))

Forward Difference Formula: f'(x_j) = (f(x_jplus1) - f(x_j)) / h

Backward Difference Formula: f'(x_j) = (f(x_j) - f(x_jminus1)) / h

Central Difference Formula: f'(x_j) = (f(x_jplus1) - f(x_jminus1)) / 2h

```python
## AN example for visualization is shown below:

## Importing the appropriate libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

h = 0.1 # step size 
x = np.arange(0, 2*np.pi, h) #the appropriate grid size
y = np.cos(x)  #function

f_diff = np.diff(y)/h   # the approximation
x_diff = x[:-1:] 
exact = -np.sin(x_diff)  #the exact solution 

plt.figure(figsize = (12, 8)) 
plt.plot(x_diff, f_diff, '-', label = 'Approximation') #plotting results
plt.plot(x_diff, exact, label = 'Solution')
plt.legend()
plt.show()
```

```python
## 20.4 Numerical Differentiation with Noise
```

Data can occasionally be tainted with noise, which causes its value to differ slightly from what it might be if it were calculated using only mathematical operations. This frequently happens in engineering because analysis of the research is inaccurate, or the data itself may be slightly altered by disturbances outside the system in question. 

```python
#An example is illustrated below:

## Importing the appropriate libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0, 2*np.pi, 0.01)  #seperation arr of numbers
omega = 100 # omega a value of 110
epsilon = 0.05 #e val
y = np.cos(x) 
y_noise = y + epsilon*np.sin(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'm-', label = 'f + noise')
plt.plot(x, y, 'b-', label = 'f')
plt.legend()
plt.show()
```
