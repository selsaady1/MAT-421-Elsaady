```python
## 17.1 Interpolation Problem Statement
```

Interpolation can be used to identify an estimate function so that y(x_i)=y_i given a particular dataset of (x_i,y_i). We may use y(x) to interpolate the function value of a new x given. y(x) is referred to be an interpolation function in this situation.

Interpolation doesnt require a particular underlying model for the given data. 

```python
import matplotlib.pyplot as plt #matplotlib library
x = [1,2,4,5] #points
y = [1,3,5,6] #points
plt.plot(x,y,'bo', 3,4.3,'ro') #graphing
plt.title('Interpolation') #graph title
plt.xlabel('x') #the x dataset labeled
plt.ylabel('y') #the y dataset labeled
plt #displaying the plot
     
```

Interpolation function can be used to estimate the data points within the range of the data set.

```python
## 17.2 Linear Interpolation
```

Linear Interpolation the point being estimated is expected to be on the line that connects both the right and left points. Considering that he x-data points are in ascending order (x_i < x < x_i+1):

Screenshot 2023-01-09 074213.png

1.) Compute the linear interpolation at x=1.25 given the dataset ---> x=[1,3,5], & y = [2,4,3].

```python
from scipy.interpolate import interp1d #importing appropriate libraries
import matplotlib.pyplot as plt #importing appropriate libraries

plt.style.use('seaborn-poster')

x = [1,3,5] # x points
y = [2,4,3] # y points
f = interp1d(x, y)
y_i = f(1.25) #prediction
print(y_i) #2.25 for the prediction

plt.plot(x, y, '-ob')
plt.plot(1.25, y_i, 'ro')
plt.title('linear Interpolation at x=1.25') #graph title
plt.xlabel('x') #the x dataset labeled
plt.ylabel('y') #the y dataset labeled
plt #displaying the plot
```

```python
## 17.3 Cubic Spline
```

Cubic spline interpolation is a set of piecewise cubic functions. Cubic functions must intersect the left&right datapoints.

To solve for CS Interpolation we have to determine the a_i, b_i, c_i, d_i values for each cubic function, then there are n-1 cubic functions for n points in whcih connect the points that need to be found. From this, we understand that there are 4(n-1) unknowns and 4(n-1) independent equations that are needed to find all the coefficients. To conduct cubic spline interpolation in Python, utilize the CubicSpline function of scipy in python of course.

An example is shown below of plotting the cubic spline interpolation of the dataset x=[1,2,3], & y=[2,4,3] through 0 to 3 interval. This is done using cubicspline.

```python
from scipy.interpolate import CubicSpline #appropriate libraries
import numpy as np #appropriate libraries
import matplotlib.pyplot as plt #appropriate libraries
plt.style.use('seaborn-poster') #style for plot

x=[1,2,3] #dataset
y = [2,4,3] #dataset
f = CubicSpline(x, y, bc_type='natural') 
x_new = np.linspace(0, 3, 100) #interval
y_new = f(x_new)

plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro') #plt
plt.title('cubicspline interpolation') #graph title
plt.xlabel('x') #the x dataset labeled
plt.ylabel('y') #the y dataset labeled
plt #displaying the plot
```
