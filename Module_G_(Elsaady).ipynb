Numerical Integration Formula= ∫[a,b] f(x)dx

We seek to estimate the integral of a function f(x) over the entire interval [a,b]. In order to achieve this, we presume that such interval has been segmented into a number grid, x, with n+1 points and h points of spacing. The count begins at x0, so there are n+1 grid points. Additionally, we suppose we have a function, f(x), which can be calculated for any grid point.

 For the function f(x) we have the integral over the interval [a,b] and that the grid consists of n+1 points with h spacing (h = (b-a)  / n). All the points on the grid are = to xi when a_0=a & x_n=b. The subinterval of this is [x_i, x_iplus1].

```python
## 21.2 Riemann's Integral
```

The summation of the areas of rectangles that are formed for every subinterval is the easiest way to approximate integrals. The elevation of the rectangle is determined by f(x) for some x in the subinterval, and the span of the rectangle is specified as xi+1xi=h. The Riemann Integral approximation is as follows:

The function result at the midpoint between xi and xi+1 is assumed by the Midpoint Rule to represent the rectangular height at every subinterval. The midpoint rule is represented as:

```python
#An example is shown below for computation and visualization:

import numpy as np #importing libraries

a = 0
b = np.pi
n = 13
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

I_riemannL = h * sum(f[:n-1])
err_riemannL = 1 - I_riemannL

I_riemannR = h * sum(f[1::])
err_riemannR = 1 - I_riemannR

I_mid = h * sum(np.sin((x[:n-1] + x[1:])/2))
err_mid = 1 - I_mid

print(I_riemannL)
print(err_riemannL)

print(I_riemannR)
print(err_riemannR)

print(I_mid)
print(err_mid)
```

In order to approximate the entire integral, the trapezoid rule slots a trapezoid through each subinterval and adds the trapezoid's areas. The trapezoid rule approximates intergrals using the following:

It is O(h3) for just a single subinterval as well as O(h2) for the entire interval because h2(f(xi+1)+f(xi)) is the estimate using the Trapezoid Rule again for integral over through the subinterval.

```python
#An example is shown below for computation and visualization:

#Here we use the trapzoid rule to approximate cosx from 0 to pi, with 13 spaced grids over exact value of 0
import numpy as np
#importing libraries

a = 0
b = np.pi
n = 13
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

I_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1])
err_trap = 0 - I_trap

#printing the results
print(I_trap)
print(err_trap)
```
