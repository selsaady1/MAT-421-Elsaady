```python

Y = dataset['products'] # We identify the Y data set as products
X = dataset['reorders'] # We identify the x data set as reorders


# We add another columns for a regression constant. This is important in this process.
X = sma.add_constant(X) 

#Here we are running our model and displaying the results.
model = sma.OLS(Y, X, missing='drop') 
DAT = model.fit()
DAT.summary()
```

## Problem III

III.  The third problem is that we want to analyze the residuals and different graphic in order to better understand the distribution [1]. This will be done through different computational procedures that will underline the key variables present.

Now, we will manually conduct the regression process in order to better understand the relationship between the variable products and reorders. This help help us extracting the equation for us to predict the future sales (reorders). This is very important because it will help us understand the possible future revenue for Instacart as a company which will only benefit them in the future. We compute this process below:

```python
#Here we form the statiscial vareiable labeling for the data sets.

y = dataset['products'] # We identify the Y data set as products
x = dataset['reorders'] # We identify the x data set as reorders
n = np.size(x) # n array

# Now we extract the means of particular variables
x_mean = np.mean(x)
y_mean = np.mean(y)
x_mean,y_mean

#Formulating the procedure of computational regression 
A = np.sum(x*y)- n*x_mean*y_mean
B = np.sum(x*x)-n*x_mean*x_mean
b1 = A/B
b0 = y_mean-b1*x_mean

#Printing the particular evaluated quantities. 
print('The slope of this mathmatical model in b1 is', b1)
print('The intercept of this mathmatical model in b0 is', b0)


#Plotting the data set with the title
plt.scatter(x,y)
plt.xlabel('The reorders')
plt.ylabel('The products')


#Formulating the prediction line of the set
y_pred = b1 * x + b0

#Finally graphing
plt.scatter(x, y, color = 'red')
plt.plot(x, y_pred, color = 'blue')
plt.xlabel('X')
plt.ylabel('y')
```

We are aware of the numerical advantage of including a constant. The difference may be seen in the addition of a "const" column of 1s to the original X matrix. Although linear regression really adds the properties of the best-fit line as well as uses the parameters to create forecasts, we may understand its purpose. A line's y-intercept as well as slope are its properties. Just a placeholder—a group of 1s to multiply the constant by—is provided in the "const" column. This demonstrates the value of including a constant. We compute this process below step by step:

```python
#Now we will add a constant.

X = sma.add_constant(X)
X.head()
```

```python
#Recall the regression results

model = sma.OLS(Y, X, missing='drop')
model_result = model.fit()
model_result.summary()
```

A significant class of methods for assessing the quality of a fitted model is residual analysis. The majority of regression analysis estimation techniques demand a properly stated regression model as well as independently occurring and uniformly distributed errors in order to be consistent, hence it is crucial to check the fundamental assumptions. Here, using the obtained mean and standard deviation, we can visualize the residuals as a kernel density plot:

```python
# The mean & standard deviation for the best fit normal distribution.
#Addressing variables and adding details
residual = DAT.resid
fig, ax = plt.subplots();
mu, std = stats.norm.fit(residual);
sns.set_palette("deep")

# Forming the cumulative residual plot
sns.histplot(x=residual, 
             ax=ax, 
             stat="density", 
             linewidth=3, 
             fill= True, 
             discrete= True, 
             color = 'pink')

#Labeling the appropriate componenets of the graphic
plt.title("The Residual Distribution") 
plt.xlabel("The Residual")
plt.ylabel("Density Of Residual") 

# Now we plot the curve and add some details for graphical accuracy
xmin, xmax = plt.xlim() 
x = np.linspace(xmin, xmax, 100000) 
p = stats.norm.pdf(x, mu, std)
sns.lineplot(x=x, 
             y=p, 
             linewidth=5, 
             color="green", 
             ax=ax)
plt.show()
```

A fit plot contrasts actual values of Y with expected quantities of the response variable.
