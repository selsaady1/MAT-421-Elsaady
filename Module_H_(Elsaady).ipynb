The relation between a function, f(x), its independent variable, x, and any degree of its derivatives is known as a differential equation. A differential equation with a one-dimensional independent variable as well as derivatives is known as an ordinary differential equation, or ODE. An ODE can be written as:

image.png

A partial differential equation, or PDE, is a differential equation of basic form where x is a vector of independent variables, x1..., xm, and the partial derivatives could be of any any degree and even with regard to any combinations of variables. A g(x) that answers the differential equation is a general solution to the equation. A differential equation typically has numerous solutions, yet these solutions are nonetheless challenging to locate. A specific solution is a p(x) that fulfills the differential equation and also has n clearly known solution values or derivative values at specific positions.

The initial value problem is the task of solving an ODE given an initial value. Initial value questions can also involve systems that develop in other regions

First-order differential equations can be solved using a variety of numerical techniques for initial value problems. We frequently need to reduce the order of the differential problem to first order classification in order for these solvers to be beneficial in solving higher order differential equations.

It is also important to note that:

The matrix form of a linear ODE is dS(t)/dt = [0, 1; -g/l, 0].

The vector S(t) represents the system's state as a function of time and this does in fact lower the order of a differential equation. A system's state is often a compilation of all the dependent variables that have an impact on how the system behaves.

```python
## 22.3 Numerical Integration Problem Statement
```

Given the condition at S(tj), we may approximate the state at S(tj+1) using the Explicit Euler Formula. We may employ this method to integrate the areas up to S(tf) starting from a specified beginning value of S0=S(t0); those S(t) values are then an estimate for the answer of the differential equation. The most straightforward and simple way to solve initial value issues is with the Explicit Euler formula. This formula is shown below:

image.png

The following below are the steps for computing s(tj) for every value tj in t:

1 Store S0 = S(t0) in an array, S.

2 Compute S(t1) = S0 + hF(t0, S0).

3 Store S1 = S(t1) in S.

4 Compute S(t2) = S1 + hF(t1, S1).

5 Store S2 = S(t2) in S.

6...

7 Compute S(t_f) = S_fminus1 + hF(t_fminus1, S_fminus1).

8 Store S_f = S(t_f) in S.

9 in an approximation of the solution to the initial value problem.

```python
#The example below illustrates solving for an ODE with an initial condition with an exact solution using a specific step size

#importing libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Define the values and boundaries
f = lambda t, s: -np.exp(-t) # ODE eq
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = 1 # Initial Condition

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
      s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'b--', label='Approx.')
plt.plot(t, np.exp(-t), 'g', label='Exact')
plt.title('Approx&Exact Solution for Simple ODE')
plt.legend(loc='upper right')
plt.show()
```
